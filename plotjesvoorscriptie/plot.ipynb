```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = "../data/IMS/2nd_test/2nd_test/2004.02.12.10.32.39"
df = pd.read_csv(dataset, sep="\t", names = ["B1", "B2", "B3", "B4"])

plt.plot(df["B2"])
plt.xlabel("Cycle (n)")
plt.ylabel("vibration/acceleration(g)")
plt.title("Vibration signal of IMS bearing 2 at 10:32:39 12-02-2004")
plt.savefig("vibration.pdf")
plt.show()
```
