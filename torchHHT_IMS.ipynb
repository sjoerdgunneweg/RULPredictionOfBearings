```python
from torchHHT import hht
import torch, os
import pandas as pd
```

```python
def imfs_IMS(fs, directory:str, dataset_num:str):

    for fileName in os.listdir(directory):
        filePath = os.path.join(directory, fileName)

        # Skip this file that gives an error, I cannot find it anywhere
        if (filePath == "data/IMS/2nd_test/2nd_test/.ipynb_checkpoints"):
            continue
            
        col_names = ['B1', 'B2', 'B3', 'B4']
        data = pd.read_csv(filePath, sep='\t', names=col_names)
    
        for col in col_names:
            tensor = torch.Tensor(data[col].values)
    
            # Use gpu if availaible to speed up
            if torch.cuda.is_available():
                device = torch.device("cuda:0")
                tensor = tensor.cuda()
    
            imfs, imfs_env, imfs_freq = hht.hilbert_huang(tensor, fs, num_imf=3)
    
            imf_string = "imfs_dataset" + str(dataset_num) + "_"  + fileName + "_" + col + ".pt"
            imf_env_string = "imfs_env_dataset"  + str(dataset_num) + "_" + fileName + "_" + col + ".pt"
            imf_freq_string = "imfs_freq_dataset"  + str(dataset_num) + "_"  + fileName + "_" + col + ".pt"

            if (dataset_num == 2):
                imf_folder = "IMFs/2nd_test"
                imf_env_folder = "IMF_ENV/2nd_test"
                imf_freq_folder = "IMF_FREQs/2nd_test"
            elif (dataset_num == 3):
                imf_folder = "IMFs/3rd_test"
                imf_env_folder = "IMF_ENVs/3rd_test"
                imf_freq_folder = "IMF_FREQs/3rd_test"
         
            torch.save(imfs, os.path.join(imf_folder, imf_string))
            torch.save(imfs_env, os.path.join(imf_env_folder, imf_env_string)) #TODO hier stond eerst imfs ipv imfs_env, voor os.path.join
            torch.save(imfs_freq, os.path.join(imf_freq_folder, imf_freq_string)) #TODO hier stond eerst imfs ipv imfs_freq, voor os.path.join 
```

```python
%%time
fs = 20000 # Sampling Frequency
directory = "data/IMS/3rd_test/4th_test/txt" # TODO, maak hier een functie van die verschilt met directory input

imfs_IMS(fs, directory, 2)
imfs_IMS(fs, directory, 3)
```
